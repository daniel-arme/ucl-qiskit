```python
from qiskit import QuantumCircuit, ClassicalRegister, QuantumRegister
from qiskit import execute, Aer
import numpy as np
from qiskit.tools.visualization import circuit_drawer
from qiskit.wrapper.jupyter import *
import matplotlib.pyplot as plt
import xlsxwriter
from qiskit import IBMQ
from qiskit.backends.ibmq import least_busy

#importing values for pair 1
import csv
method1, method2 = [],[]

with open('Book111.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in csv_reader:
        method1.append(float(row[0]))
        method2.append(float(row[1]))
```

```python
#method 1 pair 1 histogram and gaussian
import matplotlib.pyplot as plt

mu, sigma = 0.054817708, 0.072204483 # mean & standard deviation

count, bins, patches  = plt.hist(method1, 20,edgecolor='black', linewidth=1.2,facecolor='blue')

xaxis = np.linspace(min((method1)),max((method1)),1000) #more data points for smooth curve
plt.plot(xaxis, (1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (xaxis - mu)**2 / (2 * sigma**2) )), linewidth=2, color='r')
```

```python
#method 2 pair 1 histogram and gaussian

import matplotlib.pyplot as plt

mu2, sigma2 = 0.07428611, 0.118544048 # mean & standard deviation

count, bins, patches  = plt.hist(method2, 20,edgecolor='black', linewidth=1.2,facecolor='blue')

xaxis = np.linspace(min((method1)),max((method1)),1000) #more data points for smooth curve
plt.plot(xaxis, (1/(sigma2 * np.sqrt(2 * np.pi)) * np.exp( - (xaxis - mu2)**2 / (2 * sigma2**2) )), linewidth=2, color='r')
```

```python
#comparison of gaussians for method 1 and method 2 (Pair 1)

import matplotlib.pyplot as plt

mu, sigma = 0.054817708, 0.072204483 # mean & standard deviation

#count, bins, patches  = plt.hist(method1, 20,edgecolor='black', linewidth=1.2,facecolor='blue')

xaxis = np.linspace(min((method1)),max((method1)),1000) #more data points for smooth curve
plt.plot(xaxis, (1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (xaxis - mu)**2 / (2 * sigma**2) )), linewidth=2, color='r',label='Method 1')


mu2, sigma2 = 0.07428611, 0.118544048 # mean & standard deviation

#count, bins, patches  = plt.hist(method2, 20,edgecolor='black', linewidth=1.2,facecolor='blue')

xaxis = np.linspace(min((method1)),max((method1)),1000) #more data points for smooth curve
plt.plot(xaxis, (1/(sigma2 * np.sqrt(2 * np.pi)) * np.exp( - (xaxis - mu2)**2 / (2 * sigma2**2) )), linewidth=2, color='b',label='Method 2')
plt.legend()
plt.title('Pair 1: Comparison of <M> using Method 1 and Method 2')
plt.show()
```
